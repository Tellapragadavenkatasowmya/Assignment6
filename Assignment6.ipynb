```python
#importing libraries that are needed
import seaborn as sb
from sklearn import preprocessing, metrics
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pn
import numpy as nup
import matplotlib.pyplot as plot
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as sch
sb.set(style="white", color_codes=True)
import warnings
warnings.filterwarnings("ignore")
```

```python
#Reading the CSV file and finding the head 
dataframe = pn.read_csv('CC GENERAL.csv')
dataframe.head()
```

```python
dataframe.isnull().any()
```

```python
dataframe.fillna(dataframe.mean(), inplace=True)
dataframe.isnull().any()
```

```python
xline= dataframe.drop('CUST_ID', axis = 1)
print(xline)
```

```python
#Scaling is done here
scaler = StandardScaler()
scaler.fit(xline)
Xlinearray = scaler.transform(xline)
```

```python
#Here we are normalizing the data
Xnormalizedline = normalize(Xlinearray)
Xnormalizedline = pn.DataFrame(Xnormalizedline)
```

```python
#Reducing the dimensionality of the Data
pcompa= PCA(n_components = 2)
Xpca = pcompa.fit_transform(Xnormalizedline)
principalDataframe =  pn.DataFrame(data = Xpca, columns = ['principal component1', 'principal component2'])
finaldataframe= pn.concat([principalDataframe, dataframe[['TENURE']]], axis = 1)
finaldataframe.head()
```

```python
acupk2 = AgglomerativeClustering(n_clusters = 2)
plot.figure(figsize =(6, 6))
plot.scatter(principalDataframe['principal component1'], principalDataframe['principal component2'],
           c = acupk2.fit_predict(principalDataframe), cmap ='rainbow')
plot.show()
```

```python
acupk3 = AgglomerativeClustering(n_clusters = 3)
# Visualizing the clustering by showing the plots
plot.figure(figsize =(6, 6))
plot.scatter(principalDataframe['principal component1'], principalDataframe['principal component2'],
           c = acupk3.fit_predict(principalDataframe), cmap ='rainbow')
plot.show()
```

```python
acupk4 = AgglomerativeClustering(n_clusters = 4)
# Visualizing the clustering by showing the plots
plot.figure(figsize =(6, 6))
plot.scatter(principalDataframe['principal component1'], principalDataframe['principal component2'],
           c = acupk4.fit_predict(principalDataframe), cmap ='rainbow')
plot.show()
```

```python
acupk5 = AgglomerativeClustering(n_clusters = 5)
# Visualizing the clustering by showing the plots
plot.figure(figsize =(6, 6))
plot.scatter(principalDataframe['principal component1'], principalDataframe['principal component2'],
           c = acupk5.fit_predict(principalDataframe), cmap ='rainbow')
plot.show()
```

```python
acupk = [2, 3, 4, 5]
# Appending the silhouette scores of the different models to the list
sscores = []
sscores.append(silhouette_score(principalDataframe, acupk2.fit_predict(principalDataframe)))
sscores.append(silhouette_score(principalDataframe, acupk3.fit_predict(principalDataframe)))
sscores.append(silhouette_score(principalDataframe, acupk4.fit_predict(principalDataframe)))
sscores.append(silhouette_score(principalDataframe, acupk5.fit_predict(principalDataframe)))
```

```python
plot.bar(acupk, sscores)
# Plotting a bar graph to distinguish the results
plot.xlabel('The Number of clusters that we have here is ', fontsize = 10)
plot.ylabel('Silhouette scores for the AC', fontsize = 10)
plot.show()
```
